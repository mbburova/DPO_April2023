#### Все вместе:

```python
import requests
from bs4 import BeautifulSoup

url_cern = 'http://info.cern.ch/hypertext/WWW/TheProject.html' # первый веб-сайт

page = requests.get(url_cern)
#print("Status code: ", page.status_code)

page.encoding = 'utf-8'

soup = BeautifulSoup(page.text)

print(soup.prettify())
```

#### Теперь можно выбирать элементы по определенным характеристикам: тегу, классу...

```python
links = soup.find_all('a') # например, можно найти все ссылки (тег <a>)
links # список, по нему можно пройти циклом for!
```

```python
links[0] # самая первая ссылка
```

```python
links[0].text # текст ссылки
```

```python
links[0].get('href') # куда ведет ссылка
```

### Кейс: случайный вопрос из [базы вопросов ЧГК](https://db.chgk.info/)

```python
url = 'https://db.chgk.info/tour/mkm18_u' # возьмем конкретный турнир, потом можно будет добавить перебор ссылок

page = requests.get(url)
page.encoding = 'utf-8'
soup = BeautifulSoup(page.text)

print(soup.prettify())
```

```python
questions = soup.find_all("div", {"class": "question"})
num_questions = len(questions)
```

```python
questions[8].find_all('p')
```

```python
# Как выбрать случайный вопрос из списка questions?
```

```python
# решение

from random import randint

num = randint(0,num_questions)
```

```python
q = questions[num].find_all('p')[0]
print(q.text)
```

```python
# Добавим задержку между вопросом и ответом

from random import randint
import time

num = randint(0,num_questions)

q = questions[num].find_all('p')[0]
a = questions[num].find_all('p')[1]
print(q.text)

#time.sleep(60) # минута на размышление
time.sleep(10)

print(a.text)
```
